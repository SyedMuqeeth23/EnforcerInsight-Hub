**What is Groupby**

**Groupby** is a function commonly used in data analysis libraries like Pandas in Python to group data in a DataFrame based on one or more columns. It allows you to split the data into groups based on some criteria and then perform operations on these groups.

**Loading Groupby**

```python
df.head()
```

```python
df.groupby('driver_gender').search_conducted.sum()
```

```python
df.search_conducted.value_counts()
```


**What is Mapping And Data Type Casting**

**Mapping** encompasses the creation of a fresh column or altering an existing one by applying a specified function or logic. For instance, in a dataset, this process involves establishing a new column or modifying an existing one using defined rules or functions.

**Data type casting** involves changing the data type of a column in a dataset. For example, you might have a column that's read in as a string (object type in Pandas), but it actually represents numerical values.

**Loading Mapping + Data Type Casting**

```python
df.head()
```

```python
df.stop_duration.value_counts()
```

```python
df['stop_duration'] = df['stop_duration'].map( {'0-15 Min' : 7.5 , '16-30 Min' : 24 , '30+ Min' : 45 ,} )
```

```python
df
```

```python
df['stop_duration'].mean()
```
